# 1 Separation de donnees : Train et Test

```python
#x_train, x_test, y_train, y_test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 40)
```

```python
print("x-train Shape :", x_train.shape)
print("x-test Shape :", x_test.shape)
print("y-train Shape :", y_train.shape)
print("y-test Shape :", y_test.shape)
```

# 2. Construire le Modele

```python
from sklearn.linear_model import LinearRegression
```

```python
modele_lineaire = LinearRegression()
```

```python
modele_lineaire.fit(x_train, y_train)
```

```python
y_price_predits = modele_lineaire.predict(x_test)
```

```python
y_price_predits
```

```python
y_test
```

```python
resulats = pd.DataFrame(y_test , y_price_predits)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("MSE de la Regression Lineaire :", mean_squared_error(y_test, y_price_predits))
```

```python
print("r2_score de la Regression Lineaire :", r2_score(y_test, y_price_predits))
```

# 3. Visualisation des résultats

```python
plt.scatter(x_train, y_train, color='red', label='Train Data')
plt.scatter(x_test, y_test, color='blue', label='Data Test')
plt.plot(x_train, modele_lineaire.predict(x_train), color='yellow', label='Ligne de Regression')
plt.xlabel('Revenus')
plt.ylabel('Prix')
plt.title('Regression Lineaire Simple')
plt.legend()
plt.show()
```
